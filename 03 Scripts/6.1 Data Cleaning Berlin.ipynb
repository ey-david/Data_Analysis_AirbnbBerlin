# Commentary

## Empty values

Many of these values are not relevant to the analysis and can be dropped (name, description, neighborhood overview, picture_url, host_name, host_since, host_about, host_thumbnail_url,	host_picture_url, host_verifications, host_has_profile_pic,	host_identity_verified, neighbourhood,	neighbourhood_cleansed, calendar_updated)

-host_location (52739): This is a sizable amount and is relevant to the analysis. These ones will simply be excluded from that part of the analysis

-host_is_superhost (18337): This is a sizable amount and is relevant to the analysis. I will manually investigate these. Assuming these are not superhosts, I will insert a false value, or otherwise statistically impute them based on a sample.

-bathrooms can be dropped as totally empty

-bathrooms_text, maximum/minimum nights: small number, that can be safely ignored.

-beds: a small number, likely imputed by average for the room type. I actually imputed this in Excel using the "accomodates" column. While imperfect, since some beds may be doublebeds, it is reasonable.

-bedrooms: though a large number of blank rows here, looking at a sample of the data in Excel, more than half are "rooms" in the property_type column, or otherwise indicators of a one-room dwelling. All of the "rooms" can be imputed with "1", while the others can be left alone.

-review scores: probably new listings which haven't received reviews thhough a considerable number. Can be excluded from relevant analyses

license- the amount of absence is itself interesting. Looking at the data in Excel, some of them contain PAI. I will changed this that for any input will be TRUE and otherwise will be FALSE to create a boolean

## Other commentary

-latitude and longitude: need to impute a "distance_from_city_center" column using these

-amenities: this looks a bit out of scope for this analysis and will not be analyzed. However, there are probably some amenities like wifi that could make or break listings. Additionally, these can easily be counted by counting the quotation marks and then halving the count.

-source: need to investigate what effect this has, are they duplicates?

# Visualizations

```python
#minimum_nights
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df_data_wrangled_2['minimum_nights'], bins=100, edgecolor='black')


plt.title('Histogram of Minimum Nights')
plt.xlabel('Minimum Nights')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
